```python
import cv2
import numpy as np
```

```python
def framing(video):
    fr = []
    fr_pre=[]
    cap = cv2.VideoCapture(video)
    while (cap.isOpened()):
        ret,frame = cap.read()
        if ret == True:
            grayed = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            canned = cv2.Canny(grayed,320,320)
            fr.append(frame)
            fr_pre.append(canned)
        else:
            break
    cap.release()
    return fr_pre,fr
```

```python
#frames,ogframes = framing('cam_a.mp4')
frames,ogframes = framing('cam_b.mp4')
```

```python
diff = []
for i in range(0,len(frames)-1):
    print(frames[i],frames[i+1])
    diff.append(cv2.absdiff(frames[i],frames[i+1]))
```

```python
print(type(diff))
```

```python
mn = np.mean(diff)
st_d = np.std(diff)
print(mn,st_d)
```

```python
a = 4 
ts = mn + (a * st_d)
print('The threshold==>',ts)
```

```python
a_fr = []
for i in range(len(diff)):
    mn = np.mean(diff[i])
    st_d = np.std(diff[i])
    fr_ts = mn + (4*st_d)
    print(i,fr_ts)
    a_fr.append([i,fr_ts])
```

```python
imp_fr = []
for i,ac_tr in(a_fr):
    if ac_tr >= ts:
        print(i,ac_tr)
        imp_fr.append([i,ac_tr])
```

```python
key_fr = []
for i,_ in imp_fr:
    key_fr.append(ogframes[i])
    print(diff[i],i)
```

```python
for i in range(len(key_fr)):
    cv2.imshow('keys',key_fr[i])
    print("Frame no==>",i)
    k = cv2.waitKey(100) & 0xFF
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```
