```python
from gluoncv import model_zoo, data, utils
from matplotlib import pyplot as plt
import cv2
import glob
```

```python
net = model_zoo.get_model('ssd_512_resnet50_v1_voc', pretrained=True)
```

```python
imdir = '/content/drive/MyDrive/custom_dataset_keyframes/cam_b/'
files = []
files.extend(glob.glob(imdir + '*.' + 'jpg'))
```

```python
boundedbox_frames = []
count = 0
for file in files:
  x, img = data.transforms.presets.ssd.load_test(file, short=512)
  class_IDs, scores, bounding_boxes = net(x)
  ax = utils.viz.plot_bbox(img, bounding_boxes[0], scores[0], class_IDs[0], class_names=net.classes)
  plt.axis('off')
  plt.savefig('/content/drive/MyDrive/custom_dataset_boundingbox/cam_b/%d.jpg'%count, bbox_inches='tight', pad_inches = 0)
  count+=1
```
